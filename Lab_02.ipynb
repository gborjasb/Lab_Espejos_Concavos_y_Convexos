```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "d_i": [17, 14.8, 14.3, 13.7, 13.5, 12.5, 12, 12.4, 12.1, 11.9],
    "M": [-0.75, -0.55, -0.45, -0.38, -0.34, -0.3, -0.25, -0.23, -0.19, -0.15]
}

df = pd.DataFrame(data)
df
```

```python
# Linear regression
from sklearn.linear_model import LinearRegression
X = df[['d_i']]
y = df['M']
model = LinearRegression()
model.fit(X, y)

# Coefficients
slope = model.coef_[0]
intercept = model.intercept_
print(f"Slope: {slope}, Intercept: {intercept}")

# Predictions
df['M_pred'] = model.predict(X)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(df['d_i'], df['M'], color='blue', label='Data Points')
plt.plot(df['d_i'], df['M_pred'], color='red', label='Regression Line')
plt.xlabel('d_i (cm)')
plt.ylabel('M')
plt.title('Regresion Lineal de M vs d_i')
```

```python
# R^2 y RMSE
from sklearn.metrics import r2_score, mean_squared_error
r2 = r2_score(y, df['M_pred'])
rmse = np.sqrt(mean_squared_error(y, df['M_pred']))

print(f"R^2: {r2}, RMSE: {rmse}")
```
